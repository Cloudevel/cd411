# GitHub.

## Servicios de repositorios remotos de *Git*.

Debido a la popularidad de *Git*, muchos proveedores ofrecen servicios de hospedaje de repositorios remotos. Entre los más relevantes:

* [*GitHub*](https://github.com/) — El más popular, propiedad de Microsoft desde 2018.
* [*GitLab*](https://about.gitlab.com/) — Ofrece una versión *self-hosted* y una nube pública.
* [*Bitbucket*](https://bitbucket.org/) — Integrado con el ecosistema Atlassian (Jira, Confluence).
* [*Gitea*](https://gitea.io/) / [*Forgejo*](https://forgejo.org/) — Alternativas ligeras de código abierto.

Todos ofrecen hospedaje gratuito básico, pero se diferencian en sus funcionalidades avanzadas de gestión de proyectos, seguridad y DevOps.

## El servicio *GitHub*.

[*GitHub*](https://github.com/) es la plataforma de desarrollo colaborativo más utilizada del mundo, con más de 100 millones de desarrolladores registrados. Además del hospedaje de repositorios *Git*, ofrece un ecosistema completo de herramientas:

### Gestión de código
* **Repositorios** públicos y privados ilimitados.
* **Pull Requests** con revisión de código, comentarios en línea y aprobaciones.
* **Ramas protegidas** (*protected branches*) y reglas de fusión.
* **Code owners** para asignación automática de revisores.
* **GitHub Gists** — fragmentos de código compartibles.

### CI/CD y automatización
* **[GitHub Actions](https://github.com/features/actions)** — flujos de trabajo de CI/CD definidos en YAML, con miles de acciones reutilizables en el [Marketplace](https://github.com/marketplace).
* **Runners** propios (Ubuntu, Windows, macOS) o auto-hospedados (*self-hosted*).
* **Environments** con reglas de despliegue y secretos por entorno.

### Seguridad
* **[Dependabot](https://github.com/features/security)** — alertas y actualizaciones automáticas de dependencias vulnerables.
* **Secret scanning** — detección de credenciales expuestas en el código.
* **Code scanning / [CodeQL](https://codeql.github.com/)** — análisis estático para detectar vulnerabilidades.
* **[GitHub Advanced Security](https://docs.github.com/en/get-started/learning-about-github/about-github-advanced-security)** (licencia separada) — herramientas de seguridad avanzadas para repositorios privados.

### Inteligencia artificial
* **[GitHub Copilot](https://github.com/features/copilot)** — asistente de código basado en IA (modelos GPT-4o, Claude Sonnet, Gemini), disponible en el editor, en la línea de comandos y como agente autónomo (*Copilot Workspace*).

### Publicación y distribución
* **[GitHub Pages](https://pages.github.com/)** — hospedaje gratuito de sitios estáticos directamente desde un repositorio.
* **[GitHub Packages](https://github.com/features/packages)** — registro de paquetes compatible con npm, Docker, Maven, NuGet, RubyGems y Gradle.
* **[GitHub Releases](https://docs.github.com/en/repositories/releasing-projects-on-github)** — publicación de versiones con binarios adjuntos.

### Entornos de desarrollo en la nube
* **[GitHub Codespaces](https://github.com/features/codespaces)** — entornos de desarrollo completos basados en contenedores, accesibles desde el navegador o VS Code.

### Gestión de proyectos y comunidad
* **[GitHub Projects](https://github.com/features/issues)** — tableros kanban y vistas de tabla para planificación de proyectos.
* **Issues** con etiquetas, milestones y asignados.
* **[Discussions](https://github.com/features/discussions)** — foros de comunidad por repositorio u organización.
* **[GitHub Sponsors](https://github.com/sponsors)** — financiamiento directo a mantenedores de proyectos open source.

### *GitHub Enterprise*

Disponible en dos modalidades:

* **GitHub Enterprise Cloud** — alojado en la infraestructura de GitHub (Microsoft Azure). Incluye gestión centralizada de múltiples organizaciones, *Enterprise Managed Users* (EMU) para control total de identidades, cumplimiento normativo (SOC 2 Type II, FedRAMP High disponible para sector público de EE. UU.) y soporte de nivel empresarial.

* **GitHub Enterprise Server** — instalación *on-premises* o en nube privada. Permite mantener todo el código dentro de la infraestructura propia de la organización, con actualizaciones controladas. Ideal para sectores regulados (banca, defensa, salud) que no pueden usar servicios en la nube pública.

### *Forks*.

Un *fork* es una copia completa de un repositorio en la cuenta de otro usuario u organización. Es el mecanismo central del flujo de colaboración en GitHub con proyectos de código abierto:

1. Se hace *fork* del repositorio original.
2. Se trabaja en la copia propia con total libertad.
3. Se proponen los cambios al proyecto original mediante un *Pull Request*.

Los *forks* mantienen una referencia al repositorio original (*upstream*), lo que permite sincronizar cambios posteriores con:

```bash
git remote add upstream <URL-del-original>
git fetch upstream
git merge upstream/main
```

### *Pull Requests* (PR).

Un *Pull Request* es una solicitud formal para que el mantenedor de un repositorio integre cambios propuestos desde otra rama o *fork*. Es la herramienta principal de revisión de código en GitHub.

**Flujo típico:**

1. Crear una rama con los cambios (`git switch -c mi-feature`).
2. Hacer *push* de la rama al repositorio remoto.
3. Abrir un PR en GitHub indicando la rama de origen y destino.
4. Los revisores comentan, solicitan cambios o aprueban.
5. Una vez aprobado, se fusiona (*merge*, *squash* o *rebase*) en la rama destino.

**Características notables:**
* Comentarios en línea sobre cambios específicos.
* Revisiones requeridas (*required reviewers*).
* Integración con *GitHub Actions* para validaciones automáticas antes de fusionar.
* *Draft Pull Requests* para trabajo en progreso visible pero no listo para revisión.
* Vinculación automática con *Issues* (`Closes #123` en la descripción cierra el issue al fusionar).

## Actividad.

Cada participante deberá de:

* Dar de alta la llave pública SSH en su perfil.
* Obtener un *Personal Acess Token* (*PAT*) para operaciones con sus repositorios.
* Hacer un *fork* del repostorio del capacitador.
* Añadir desde *Github* un archivo en el repositorio creado con el *fork*.
* Solicitar un *pull request* del repositorio del participante con el reposiotrio original.
